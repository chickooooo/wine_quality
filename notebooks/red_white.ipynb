### Data Collection

```python
# training data
train_data = pd.read_csv("../data/train.csv")
train_data.drop(['Id'], axis=1, inplace=True)
train_data.head()
```

```python
# red wine dataset
red_data = pd.read_csv("../red_white_dataset/winequality-red.csv", delimiter=';')
red_data.head()
```

```python
# white wine dataset
white_data = pd.read_csv("../red_white_dataset/winequality-white.csv", delimiter=';')
white_data.head()
```

<br>
<br>

### Data Preparation

```python
from sklearn.model_selection import train_test_split
```

```python
# creating validation set
x_train, x_val = train_test_split(train_data, test_size=0.3, shuffle=True, stratify=train_data.quality)
```

```python
# inspecting shape
print(x_train.shape)
print(x_val.shape)
```

```python
# combining data
total_data = pd.concat([x_train, red_data, white_data], ignore_index=True)
total_data.head()
```

```python
# shape
total_data.shape
```

```python
# removign duplicate values
total_data.drop_duplicates(inplace=True)
```

```python
# updated shape
total_data.shape
```

```python
# inspecting final data
total_data.head()
```

### AutoML

```python
# quadratic weighted kappa
def quad_kappa(y, y_pred):
    return cohen_kappa_score(y, y_pred, weights='quadratic')
```

<br>
<br>

```python
# creating environment
total_wine_quality = pcc.setup(data=total_data, target='quality', use_gpu=True)
```

```python
# adding custom metric
pcc.add_metric('quad_kappa', 'Quad Kappa', quad_kappa)
```

```python
# comparing models
pcc.compare_models()
```

```python
# getting top 4 models
top_4 = pcc.compare_models(sort='Quad Kappa', include=['rf', 'qda', 'lightgbm', 'catboost'], n_select=4)
```

```python
# blending models
blender = pcc.blend_models(top_4, optimize='Quad Kappa')
```
